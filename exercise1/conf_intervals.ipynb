```r
# Bootstrapping

set.seed(1)
# get a random distribution 
x = rnorm(n=50, mean=10, sd=2)
# find the standard error 
se_x = sqrt(var(x)/length(x))

out = NULL
for(i in 1:1000){
  # 50 values (sampling one dataset)
  sample = sample(x, replace=TRUE)
  # calculating the mean of the dataset
  out[i] = mean(sample)
}

# out now contains the sampling distribution
# plot it on a histogram
hist(out, las=1, main="")
```

```r
sd(out)
```

```r
se_x

# the standard deviation of the bootstrapped dataset
# gives an approximation of the standard error
```

```r
quantile(out, c(0.025, 0.975))
```

```r
qnorm(c(0.025, 0.975))
```

We could have generated a confidence interval with 1.96SE. This works because for a normal distribution, 95% of the probability lies within ~1.96 standard deviations from the mean, (That is, the 2.5th percentile = -1.96, the 97.5th percentile = +1.96) so we use that property to build a 95% confidence interval for the mean.

```r
# quantiles of normal distribution
qnorm(c(0.025, 0.975))
```

```r
# start of 95% confidence interval
mean(x) - 1.96*se_x
```

```r
# end of 95% confidence interval
mean(x) + 1.96*se_x
```

Simple analogy

Imagine you want to guess the average height of daisies in a meadow, but you only have 10 daisies.

You resample your 10 daisies with replacement 1000 times, calculate the average each time  you now have a cloud of plausible means.

The 2.5th and 97.5th percentiles of this cloud are your empirical 95% CI.

You dont need normality  the resampling itself estimates the uncertainty.
