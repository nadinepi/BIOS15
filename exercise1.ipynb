```r
calculate_cv <- function(x) {
  cv <- sd(x) / mean(x)
  return(cv)
}
```

```r
set.seed(8)
x <- rnorm(n=20, mean=5, sd=1)

cv_values <- numeric(1000)
for (i in 1:1000) {
  sample <- sample(x, replace=TRUE)
  cv_values[i] <- calculate_cv(sample)
}

hist(cv_values)
```

```r
quantile(cv_values, c(0.025, 0.975))
```
